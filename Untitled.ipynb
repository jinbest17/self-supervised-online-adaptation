```python
import csv
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gas_accuracy = []
with open('gas.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        for element in row:    
            gas_accuracy.append(element)
```

```python
gas_accuracy = np.array(gas_accuracy,dtype=np.float64).reshape(6,9)
```

```python
gas_accuracy_mean = np.mean(gas_accuracy,axis=0).reshape(9,)
```

```python
lssvm_accuracy = [0.8336012861736335,0.5958385876418664,0.2670807453416149,0.20304568527918782,0.2917391304347826,0.27041239966786607,0.11224489795918367,0.2148936170212766,0.3591666666666667]
```

```python
ot_accuracy = [0.6889067524115756,0.3329129886506936,0.14285714285714285,0.1065989847715736,0.17173913043478262,0.2839745363963465,0.3231292517006803,0.06382978723404255,0.2038888888888889]
```

```python
np.divide(gas_accuracy_mean,lssvm_accuracy)
```

```python
np.divide(gas_accuracy_mean,ot_accuracy)
```

```python
df = pd.DataFrame()
df['Accuracy'] = np.concatenate((gas_accuracy_mean,lssvm_accuracy,ot_accuracy))
batch = []
for i in range(2,11):
    batch.append(i)
df['Batch #'] = np.concatenate((batch,batch,batch))
df['Model'] = np.concatenate((['SelfTrain']*9,['uPAC']*9,['OT']*9))
```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 34})
fig = plt.figure(figsize=(20,10))

ax = sns.barplot(x="Batch #", y="Accuracy", hue="Model", data=df)
plt.grid()
plt.show()
```
