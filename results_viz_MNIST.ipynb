```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
saved = pickle.load(open("mnist_results.p","rb"))
```

```python
print(saved
     )
```

```python
count = 0
for i in range(len(saved['results_rotate'])):
    if saved['results_rotate'][i] == saved['true_label'][i]:
        count+=1
print(count /len(saved['results_rotate']) )
```

```python
count = 0
for i in range(len(saved['results_perm'])):
    if saved['results_perm'][i] == saved['true_label'][i]:
        count+=1
print(count /len(saved['results_rotate']) )
```

```python
count = 0
for i in range(len(saved['results_perm'])):
    if saved['results_noise'][i] == saved['true_label'][i]:
        count+=1
print(count /len(saved['results_rotate']) )
```

```python
results= saved['results_noise']
countnum = 0
results_100 = []
stepSize = 100
print(len(results))


for step in range(0,32):
  countnum = 0
  
  for i in range(stepSize):
      if results[i +step*stepSize] != saved['true_label'][i+step*stepSize]:
          countnum+=1
  print(1-countnum/stepSize)
  results_100.append(1-countnum/stepSize)
indices = np.arange(32)
indices = indices * 100
print(indices)
plt.plot( indices,results_100)
plt.ylim([0, 1])
plt.show()
```
